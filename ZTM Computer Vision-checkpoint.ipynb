## Online images

```python
import requests
from io import BytesIO

def url_image_to_cnn_binary(model, url):
    response = requests.get(url)
    response.raise_for_status()
    online_pic = Image.open(BytesIO(response.content)).convert("RGB")
    img_resize = online_pic.resize(size=(224,224))
    np_img = np.array(img_resize)
    np_img_normalized = np_img/255
    tf_image = tf.expand_dims(np_img_normalized, axis=0)
    img_type = model.predict(tf_image)
    if img_type > 0.75:
        return "Stake"
    elif img_type > 0.25:
        return "Hard to tell"
    else:
        return "Pizza"
```

# 8. Multiclass Crassification

```python
train_dir = "C:/Users/iosif/Desktop/Python/Tensorflow/10_food_classes_all_data/train"
test_dir = "C:/Users/iosif/Desktop/Python/Tensorflow/10_food_classes_all_data/test"
#------------------------------------
train_datagen = ImageDataGenerator(rescale=(1/255),
                                   rotation_range=0.2,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   zoom_range=0.2,
                                   shear_range=0.2,
                                  horizontal_flip=True,
                                   vertical_flip=True)

test_datagen = ImageDataGenerator(rescale=(1/255),
                                   rotation_range=0.2,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   zoom_range=0.2,
                                   shear_range=0.2,
                                  horizontal_flip=True,
                                   vertical_flip=True)
#------------------------------------
train_dataset = train_datagen.flow_from_directory(directory = train_dir,
                                                  target_size = (224,224))

test_dataset = test_datagen.flow_from_directory(directory = test_dir,target_size = (224,224))
```

```python
food_names = train_dataset.class_indices
food_names

names_of_food = {v: k for k, v in food_names.items()}
names_of_food
```

```python
images, labels=next(train_dataset)

for i in range(0,9):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i])
    plt.title(names_of_food[np.argmax(labels[i])])
    plt.axis("off")
plt.tight_layout
plt.show()
```

```python
food_cat_viz_1 = Sequential([
    Conv2D(filters=10, kernel_size=3, activation="relu", input_shape=(224,224,3)),
    Conv2D(filters=10, kernel_size=3, activation="relu"),
    MaxPool2D(),
    Conv2D(10,3,activation="relu"),
    Conv2D(10,3,activation="relu"),
    MaxPool2D(),
    Flatten(),
    Dense(10, activation="softmax")
])

food_cat_viz_1.compile(loss = tf.keras.losses.CategoricalCrossentropy(),
                      optimizer = "Adam",
                      metrics = "accuracy")

hist_2 = food_cat_viz_1.fit(train_dataset,
                           epochs=5,
                           steps_per_epoch=len(train_dataset),
                           validation_data=test_dataset,
                           validation_steps=len(test_dataset))
```

```python
pd.DataFrame(hist_2.history).plot()
```

```python
food_cat_viz_2 = Sequential([
    Conv2D(10,3,activation="relu", input_shape=(224,224,3)),
    Conv2D(10,3,activation="relu"),
    MaxPool2D(),
    Conv2D(10,3,activation="relu"),
    Conv2D(10,3,activation="relu"),
    MaxPool2D(),
    Flatten(),
    Dense(10, activation="softmax")
])

food_cat_viz_2.compile(loss= "categorical_crossentropy",
                      optimizer = "Adam",
                      metrics = "accuracy")

hist_3 = food_cat_viz_2.fit(train_dataset,
                           epochs = 10,
                           validation_data=test_dataset,
                           validation_steps=len(test_dataset),
                           steps_per_epoch=len(train_dataset))
```
