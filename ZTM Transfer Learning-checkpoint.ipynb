```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import os
```

```python
train_dir = "C:/Users/iosif/Desktop/Python/Tensorflow/10_food_classes_10_percent/train"
test_dir = "C:/Users/iosif/Desktop/Python/Tensorflow/10_food_classes_10_percent/test"
#----------------------------------------------------
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1/255)

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1/255)
#----------------------------------------------------
train_data = train_datagen.flow_from_directory(directory=train_dir,
                                                target_size=(224,224))

test_data = test_datagen.flow_from_directory(directory=test_dir,
                                             target_size=(224,224))
```

```python
def cerate_tensorboard_callback(dir_name, experiment_name):
    log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
    return tensorboard_callback
```

```python
# Clear the top layer
base_model = tf.keras.applications.ResNet50(weights="imagenet", include_top=False, input_shape = (224,224,3))
# Preserve the weights
base_model.trainable = False
# Add custom layers
x = base_model.output
# Polling to redue dimentions
x = tf.keras.layers.GlobalAveragePooling2D()(x)
# Output
last_line = tf.keras.layers.Dense(10, activation="softmax")(x)
#-------------------------------
# Build
my_resnet = tf.keras.Model(inputs=base_model.inputs, outputs=last_line)
# Compile
my_resnet.compile(loss="categorical_crossentropy",
                 optimizer="Adam",
                 metrics="accuracy")
# Fit
hist_resnet = my_resnet.fit(x=train_data,
             epochs=5,
             validation_data=test_data,
             steps_per_epoch=len(train_data),
             validation_steps=len(test_data))
```

```python
pd.DataFrame(hist_resnet.history).plot()
```

```python
my_resnet.evaluate(test_data)
```
